```python
import numpy as np
```

```python
# 1-Criando os Dados
nomes = np.array(['Alice', 'Bob', 'Charlie', 'David', 'Eva'])
notas_mate = np.array([55, 67, 75, 48, 90])
notas_fis = np.array([88, 76, 95, 64, 70])
```

```python
# 2-Calculando médias das notas
media = np.mean(np.column_stack((notas_mate,notas_fis)), axis=1)
media
```

```python
# 3- Identificando alunos aprovados
alunos_aprovados = np.where(media > 60)
alunos_aprovados
```

```python
# 4- Organizando dados em uma tabela
dados_tabela = np.column_stack((nomes, notas_mate, notas_fis, media))
dados_tabela
```

```python
# 5- Exibindo os alunos aprovados
for i in alunos_aprovados[0]:
        print(f"{dados_tabela[i][0]} - Média:{media[i]:.2f}")
```

```python
# 6- Divindo alunos em grupos
grupos = np.array_split(nomes, 3)
for i, value in enumerate(grupos, 1):
    print(f"Grupo {i}: {value}")
```
