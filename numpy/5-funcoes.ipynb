```python
import numpy as np
```

```python
array = np.array([1, 2, 3, 4, 5])
array
```

```python
# 1- Adição
soma = np.add(array, 10)
soma
```

```python
# 2- Subtração
subtracao = np.subtract(array, 5)
subtracao
```

```python
# 3- Multiplicação
multiplicar = np.multiply(array, 2)
multiplicar
```

```python
# 4- Divisão
divisao = np.divide( array, 2)
divisao
```

```python
# 5- Funções Trigonometricas
seno = np.sin( array)
seno
```

```python
conseno = np.cos(array)
conseno
```

```python
# 6- Funções Estatisticas
media = np.mean(array)
media
```

```python
desvio_padrao = np.std(array)
desvio_padrao
```

```python
# 7- Funções de comparação
maior_que_3 = array > 3
maior_que_3
```

```python
# 8- Combinando Funcões
resultado = np.sqrt(array) + np.power(array, 2)
resultado
```

```python
# 9- Funções Exponenciais
exponencial = np.exp(array)
exponencial
```

```python
# 10- Função de Logaritmo
logaritmo = np.log(array)
logaritmo
```

```python
# 11- Funçao de Arredondamento
arredondado = np.round(array /2)
arredondado
```

```python
# 12 - Função de potencia
potencia = np.power(array, 3)
potencia
```
